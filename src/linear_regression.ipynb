```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 1.5*x

pts = np.arange(start=1, stop=10, step=0.4)
y = f(pts)
plt.plot(pts, y, c='y')
plt.grid(True)
```

```python
dataset = np.array([h + np.random.normal() for h in y])

plt.grid(True)
plt.plot(pts, y, c='y')
plt.scatter(pts, dataset, c='b', marker='x')
plt.legend(['original function', 'training data'])
```

```python
# hypothesis function
def h(x, w=1):
    return w*x

def cost_function(w):
    cost = 0
    for i in range(23):
        cost += (h(pts[i], w) - y[i])**2
    return 0.5*cost

# returns weight after performing stochastic gradient descent
def gradient_descent(w):
    alpha = 0.001 # learning rate 
    res = 0

    for i in range(23):
        res += pts[i] * (y[i] - h(pts[i], w)) 

    return w + alpha * res

# assuming initital weight as 1 
w = 1 

while cost_function(w) > 1:
    w = gradient_descent(w)
```

```python
# calculating error
mean_square_error = 0
for i in range(23):
    mean_square_error += (y[i] - w*pts[i])**2

mean_square_error = mean_square_error / 23
print(f'the hypothesis function comes out to be: {round(w, 2)}x')
print(f'mean square error: {round(mean_square_error, 3)}%')
```
